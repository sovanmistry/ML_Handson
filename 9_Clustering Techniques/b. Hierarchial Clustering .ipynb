## Hierarchial Clustering 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
#load the dataset 

iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data.head()
```

```python
## Unsupervised technique needs Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#Scaled data from Feature Engineering

X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled.shape
```

## Applying PCA - converting 4 feature into 2 feature 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python

pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1])
```

```python
pca_scaled
```

```python
## Agglomerative Clustering 

import scipy.cluster.hierarchy as sc

sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
```

## Agglomerative Clustering to implement Hierarchial Clustering

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')

#n_clusters =2 from above picture
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
#only 2 class
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c = cluster.labels_)
```

## silhouette_score

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficient = []
```

```python
for k in range (2,11):
    agglo = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_coefficient.append(score)
```

```python
plt.plot(range(2,11), silhouette_coefficient)
plt.xticks(range(2,11))
plt.show()
```
